**Algorithm**

According to the article "Movie Recommender System using Single Value Decomposition and K-means Clustering," the K-means Clustering algorithm was used. Its purpose is to divide the observations into k clusters according to theirs centers of gravity.

The algorithm has 4 main steps:


*  Step 1: Choose the number of clusters k. 
*  Step 2: Select k random points from the data as centroids.
*  Step 3: Assign all the points to the closest cluster centroid.
*  Step 4: Recompute the centroids of newly formed clusters.
*  Step 5: Repeat steps 3 and 4.

The function partition the given N dataset into k disjoint subsets Sj which consist of Nj data points as close as possible to the other points on the basis of similarity measures. Each partition contains Nj data points for the given Cj centroids. In order to make a good partition the distance between all the data points with the centroids should be minimized.

**Clusters-SSE Knee-Point (SSE Derivation)**

The Sum of squared errors was calculated for all clusters. As the number of clusters increases the smaller the distance between the group members (SSE).

In each data set, 4 knees of a curve were identified (minimum points) that constitute a potential division into clusters.
It was decided to perform the examination on 7 clusters for the two datasets for the following reasons:

* The minimum points are the same in the two data sets, which will facilitate future comparison.
* Dividing clusters into a number exceeding 7 clusters presented an OVERFITTING issue which will be described later.

**Determining the number of clusters in a data set** 

If we divide the dataset into 7 clusters we will get one cluster with one observation. If we divide the data set into 10 clusters we will get the following result - in clusters 2 and 8 there will be only one observation. These cases are cases of overfitting- the model is over-adapted to the collection of data and therefore the execution of the prediction will be less successful.

Increasing the number of clusters will receive more cases of overfitting  - there will be more clusters with individual observations.
Conclusion: The optimal division is - 7 clusters.

The RMSE of each cluster was calculated and presented in a table. 

* First step- Square Error calculation (in Excel): substraction of mean value of each dimension within the clusters from mean values of dimension within the cluster X in power of 2.

**Review of the Literature**

The literature deals a lot with the methods that serve as the basis for a film recommendation system. The larger the database, the more complex the techniques are required to respond quickly. In addition, the nature of the model requires constant updating.

There are two main approaches on which the film recommendation can be based:
* Approach 1: Based on the characteristics/features of the film - CONTENT BASED.
* Approach 2: Based on a rating of other users - COLLABORATIVE.
